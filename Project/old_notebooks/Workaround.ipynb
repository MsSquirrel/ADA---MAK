```python
%matplotlib inline
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from pandas.io.json import json_normalize
import gzip
sns.set()
```

```python
def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield eval(l)

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')

df_review = getDF('data/reviews_Cell_Phones_and_Accessories.json.gz')
df_meta = getDF('data/meta_Cell_Phones_and_Accessories.json.gz')
```

```python
categories = df_meta.categories
```

```python
categories = []
[categories.append(i) for i in df_meta.categories if not categories.count(i)]
categories = [item[0] for item in df_meta.categories.values]
categories_df = pd.DataFrame(categories)
categories_df.columns = [str(col) + '_cat' for col in categories_df.columns]
df_meta = pd.concat([df_meta, categories_df], axis=1)
```

```python
df_phones = df_meta.loc[(df_meta['1_cat'] == 'Cell Phones') & (df_meta['2_cat'] == 'Unlocked Cell Phones')]
df_phones = df_phones[(pd.notnull(df_phones['brand'])) & (df_phones['brand'] != '')]
```

```python
df_final= pd.merge(df_review, df_phones, left_on='asin', right_on='asin')
```
