# Solution

### Reasoning


Our understanding of the recent date is previous year - 2016, as it is long enough time interval to get some real statistics about the unemployment. Other options that we considered were one month and half a year, but we think using this approach the results would be influenced by mountly migrations. For example more people are traveling during the winter and summer months as these are the holiday seasons, and are less likely to search for a job during these months. 


### Load dataset
To begin, we download a dataset on the [Eurostat](http://ec.europa.eu/eurostat/data/database) website by using the search bar and looking for the term *Unemployment*. We have chosen the dataset called "Total unemployment rate" (*tsdec450*) that contains unemployment rates per year, for most of the european countries.

 

```python
total_unemployment_file_path = 'data/tsdec450.tsv'

# load the data
def load_data(file_name):
    df = pd.DataFrame()
    df = df.append( pd.read_csv(file_name, sep="\t" , header=0))
    # clean the column names
    df.columns = [i.split(" ")[0] for i in df.columns]
    df.columns = [i.split("\\")[0] for i in df.columns]
    new_col = df["age,unit,sex,geo"].str.split(",")
    # from one column with all data create two separate columns that are important to us
    df["State"] = [pi[3] for pi in new_col]
    df["Unit"] = [pi[1] for pi in new_col]
    
    return df

df = load_data(total_unemployment_file_path)
df.head()
```

We found some mismatching in json geo-coordinate files, and country codes for two countries. Because of this we replace this codes in the dataframe in order to be consistent. Measurement unit's code for the metric that we want to present is PC_ACT, so we filter by this value.

```python
#Pre-processing : keep only useful data
df['2016'] = df['2016'].astype(float)
df = df.query('Unit == \'PC_ACT\'')

# Put the same country code in the dataset and in json geo-coordinate files
df = df.replace('UK', 'GB')
df = df.replace('EL', 'GR')
```

```python
df = df[['State', '2016']]
df.head()
```

### Define classification function
According to [GISGeography.com](http://gisgeography.com/choropleth-maps-data-classification/), we learn that selecting a data classification method is very important. It impacts very much our understanding of the data and thus it is important to find the right classification. We decided to code and test two of them:
- **Equal Interval Data Classification** as *equal* : It divides the classes into equal groups of same interval length.
- **Quantile classification** as *quantile*: It arranges the classes to have the same quantity inside.

```python
# define functions for classifying the unemployment
def define_threshold(df, column_name, classification='equal', classes=4):
    # in case of 'equal' classification divide the groups into the intervals
    # of the same length
    if classification == 'equal':
        min_val = df[column_name].min()
        max_val = df[column_name].max()
        bins = []
        bins.append(min_val - 0.001)
        for i in np.arange(1, classes):
            bins.append(((max_val - min_val)* i/classes) + min_val)
        bins.append(max_val + 0.001)
        return bins
    
    # in case of 'quantile' classification divide the classes 
    # in such way that they have the same quantity inside
    if classification == 'quantile':
        out, bins = pd.qcut(df[column_name], classes, retbins=True)
        bins += 0.001
        return bins.tolist()
```

### Plot european map


We are plotting the map of percentage of unemployed people in european countries according to statistics for year 2016. This dataset does not provide us the data for each european country, so we had to think how to present this in understandable manner on the map. We choose to leave these countries colored the same way as the rest of the world, for which we do not show any statistics. This way intuitively should be clear that the data for these countries is missing. 

```python
geojson_path = 'topojson/europe.topojson.json'
topojson_path = json.load(open(geojson_path))

# Countries that were missing in the dataset are in the same color
# as the rest of the world for which we do not show any data
topojson_path['objects']['europe']['geometries'] = [
    feature for feature 
    in topojson_path['objects']['europe']['geometries'] 
    if (feature['id'] in df['State'].values) ]
```

#### Quantile approach

```python
# Draw the map with our threshold classification function
# We use quantile approach here
eu_map = folium.Map([52.3, 9.7], tiles='cartodbpositron', zoom_start=3.6)
eu_map.choropleth(geo_data=topojson_path , data=df, topojson='objects.europe',
             columns=['State', '2016'],
             key_on='feature.id',
             fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.2,
             threshold_scale=define_threshold(df, '2016', classification='quantile',classes=4),
             legend_name='Percentage of unemployment (%)')

eu_map.save('maps/unemployment_europe_quantile_classfication.html')
eu_map
```

## 2.2 - Unemployment rates : various interpretations

### Compute new "unemployment" ratios

As said in the explanation of this task, The Swiss Confederation defines previousy calculated and presented unemployment rates as the number of people looking for a job divided by the size of the active population (scaled by 100). This is surely a valid choice, but as we discussed one could argue for a different categorization.

Another possible categorization is to calculate the unemployment rate only taking into account people who do not have a job. So excluding people who have job, but are searching for another, we get little bit different results. 

First of all, we used the known unemployment rate and number of all jobseekers to calculate the exact population. We used simple math for this. We get the population by diving the number of all jobseekers with unemployment rate (this is 1% of all population), and then we multiply it by 100 to get 100% or all population. 

After this, we calculate unemployment rate using another approach, taking into account only people who are without a job. We calculate proportion of this people over all population, and then multiply by 100 to get percentage instead of proportion.

```python
switzerland_df['population'] = (
    switzerland_df['all_jobseekers'] / switzerland_df['unemployment_rate']) * 100

switzerland_df['unemployed_jobseekers_unemployment_rate'] = (
    switzerland_df['unemployed_jobseekers'] / switzerland_df['population'] ) * 100

switzerland_df.head()
```

```python
# we will use the values from prevoius map to decide on threshold_scale
# This way both of the maps will have the same scale for colours and 
# we can use coloring to understand the difference

def define_threshold_switzerland(df, column_name , classes=4 ):
    min_val = df['unemployment_rate'].values.min()
    max_val = df['unemployment_rate'].values.max()
    bins = []
    bins.append(min_val - 0.001)
    for i in np.arange(1, classes):
        bins.append(((max_val - min_val)* i/classes) + min_val)
    bins.append(max_val + 0.001)
    return bins
```

## 3.2 - Nationality and age groups

Now let's refine the analysis by adding the differences between age groups. As you may have guessed it is nearly impossible to plot so many variables on a map. Make a bar plot, which is a better suited visualization tool for this type of multivariate data.

### Assumptions
We don't achieve to get unemployment rates for all classes (age and nationality).<br>
As a matter of fact, we decided to show the distribution between all the subclasses using one final horizontal barplot.

```python
def define_age_classes(x):
    if(x == '1'):
        return ' 15 - 24 years'
    elif(x == '2'):
        return ' 25 - 49 years'
    else:
        return ' 50+ years'
```

```python
switzerland_age_groups_path = 'data/unemployment_nationality_age.csv'

#Some preprocessing 
switzerland_age_df = pd.read_csv(open(switzerland_age_groups_path,
                                      encoding='utf-16'),
                                      header=[1,2],
                                      thousands='\'')

#Remove total rows as we don't need it
switzerland_age_df = switzerland_age_df[~switzerland_age_df.isin(['Total'])]
switzerland_age_df = switzerland_age_df.drop(['Mesures', 'Unnamed: 3_level_1'],
                                             axis=1, level=1) .dropna()
                                           
switzerland_age_df.rename(
                columns={'Classes d\'Ã¢ge 15-24, 25-49, 50 ans et plus': 'classes'},
                inplace=True)


switzerland_age_df.columns = switzerland_age_df.columns.droplevel(1)
switzerland_age_df['Total'] = np.round(switzerland_age_df.mean(axis=1).astype('int64'))
switzerland_age_df['classes'] = switzerland_age_df['classes'].apply(
                                lambda x: define_age_classes(x))

switzerland_age_df['Nationality + age'] = switzerland_age_df['NationalitÃ©'] + ' ' + switzerland_age_df['classes']

# Use pivot function for horizontal bar plot
switzerland_age_df = switzerland_age_df.pivot(index='Canton',
                                              columns='Nationality + age',
                                              values='Total')


switzerland_age_df['total'] = [0 for i in range(len(switzerland_age_df))]

for p in switzerland_age_df.columns[0:len(switzerland_age_df.columns)-1]:
    switzerland_age_df['total'] += switzerland_age_df[p]
    
for p in switzerland_age_df.columns[0:len(switzerland_age_df.columns)-1]:
    switzerland_age_df[p] = switzerland_age_df[p]/switzerland_age_df['total']
    
switzerland_age_df.drop('total', axis=1, inplace=True)

switzerland_age_df.head()
```

### Analysis on the distribution

```python
fig, ax = plt.subplots(figsize=(14,8))
switzerland_age_df.plot.barh(stacked=True, grid=True, width=0.7, ax=ax)
plt.xlabel('Proportion of registered unemployed people',fontsize=30)
plt.ylabel('Canton', fontsize=30)
plt.yticks(fontsize=20)
plt.legend(fontsize=20, shadow=True)
plt.show()
```

```python
# TODO: we can also use box plots to show the distribution of the values among each columns
switzerland_age_df.plot.box(figsize=(16,8))
```

### Observations
So here we see that the age category from 25 to 49 years is the most touched by unemployment, regardless of nationality. Regardless of nationality, we also see that the age category for 15 to 24 years is the least touched by unemployment.

Looking at the boxes, we see that the difference between swiss and foreign unemployed is bigger when we look at age category from 50+ years, where foreign people are way less touched by unemployment.

It is also interesting to see that for 2 categories (15 - 24 years and 50+ years), foreign people are less unemployed than swiss people, but when we compute unemployment for swiss and foreign people in task 2, foreign people were more unemployed than swiss people. So we can suppose that there is more foreign people from age category from 25 to 49 years than from the 2 others categories, so they take the unemployment rate up (as they are the most unemployed category).

## Methodology

In order to see nicely the three Rostigraben regions on the map, we plot cantons with different bording colors, according if people speak French (FR), German (DE) or Italian (IT).

```python
def rostigraben_colors(feature_id):
    IT = ['TI']
    FR = ['GE', 'VD', 'FR', 'NE', 'VS', 'JU']
    
    if feature_id in IT:
            return '#4daf4a';
    elif feature_id in FR:
        return '#377eb8'
    else:
        return '#e41a1c'
```

```python
rostigraben_map = folium.Map([46.8001, 8.1958], tiles='cartodbpositron', zoom_start=8)

# Layer 1: unemployment rate
rostigraben_map.choropleth(geo_data=switzerland_topojson, data=switzerland_df,
             columns=['code', 'unemployment_rate'],
             key_on='feature.id',
             fill_color='YlOrRd', fill_opacity=0.3, line_opacity=0,
             legend_name='Unemployment rate - Swiss population (%)',
             name='Unemployment rates - Swiss population',
             topojson='objects.cantons',
             threshold_scale=define_threshold(switzerland_suisses_df,'unemployment_rate', 'equal',5))

# Layer 2: cantons borders according to rostigraben regions
folium.TopoJson(
    switzerland_topojson,
    name = 'Languages spoken according to Rostigraben',
    object_path ='objects.cantons',
    style_function=lambda feature: {
        'fill_color': 'white',
        'fillOpacity': 0,
        'weight' : 3,
        'color': rostigraben_colors(feature['id']),
        'opacity': 1  
    }
).add_to(rostigraben_map)

folium.LayerControl(collapsed=False).add_to(rostigraben_map)
rostigraben_map.save('maps/unemployment_switzerland_rostigraben.html')
rostigraben_map
```
